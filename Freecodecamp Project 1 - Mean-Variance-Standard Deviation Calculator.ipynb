# Mean-Variance-Standard Deviation Calculator

---

```python
#import relavent packages
import numpy as np
import pandas as pd
```

```python
def calculate(list):
    
    if (len(list)!= 9):
        raise ValueError("List must contain nine numbers.")
        
    new_list = np.array(list)
  
    cal_mean_rows = ([new_list[[0,1,2]].mean(), new_list[[3,4,5]].mean(), new_list[[6,7,8]].mean()])
    cal_mean_cols = ([new_list[[0,3,6]].mean(), new_list[[1,4,7]].mean(), new_list[[2,5,8]].mean()])
    cal_variance_rows = ([new_list[[0,1,2]].var(), new_list[[3,4,5]].var(), new_list[[6,7,8]].var()])
    cal_variance_cols = ([new_list[[0,3,6]].var(), new_list[[1,4,7]].var(), new_list[[2,5,8]].var()])
    cal_std_devi_rows = ([new_list[[0,1,2]].std(), new_list[[3,4,5]].std(), new_list[[6,7,8]].std()])
    cal_std_devi_cols = ([new_list[[0,3,6]].std(), new_list[[1,4,7]].std(), new_list[[2,5,8]].std()])
    cal_max_rows = ([new_list[[0,1,2]].max(), new_list[[3,4,5]].max(), new_list[[6,7,8]].max()])
    cal_max_cols = ([new_list[[0,3,6]].max(), new_list[[1,4,7]].max(), new_list[[2,5,8]].max()])
    cal_min_rows = ([new_list[[0,1,2]].min(), new_list[[3,4,5]].min(), new_list[[6,7,8]].min()])
    cal_min_cols = ([new_list[[0,3,6]].min(), new_list[[1,4,7]].min(), new_list[[2,5,8]].min()])
    cal_sum_rows = ([new_list[[0,1,2]].sum(), new_list[[3,4,5]].sum(), new_list[[6,7,8]].sum()])
    cal_sum_cols = ([new_list[[0,3,6]].sum(), new_list[[1,4,7]].sum(), new_list[[2,5,8]].sum()])
    
    return {'mean': [cal_mean_rows, cal_mean_cols, new_list.mean()],
            'variance': [cal_variance_rows, cal_variance_cols, new_list.var()],
            'standard deviation': [cal_std_devi_rows, cal_std_devi_cols, new_list.std()],
            'max': [cal_max_rows, cal_max_cols, new_list.max()],
            'min': [cal_min_rows, cal_min_cols, new_list.min()],
            'sum': [cal_sum_rows, cal_sum_cols, new_list.sum()]}
```
